```python
import pandas as pd

df = pd.read_csv("news_comment_BTS.csv", encoding="utf-8")
df.info()
```

```python
# 불필요한 문자 제거
# 이건 데이터프레임이기 때문에 replace 로 제거한다.
df["reply"] = df["reply"].str.replace("[^가-힣]", " ", regex=True)
df["reply"].head()
```

```python
import konlpy

kkam = konlpy.tag.Kkma()
kkam
```

```python
# 명사 추출
nouns = df["reply"].apply(kkam.nouns)
nouns
```

```python
# 한 행에 한 단어가 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
# 데이터 프레임 만들기
df_word = pd.DataFrame(
    {
        "word": nouns,
    }
)

# 글자 수 추가
df_word["count"] = df_word["word"].str.len()

# 두 글자 이상만 남기기
df_word = df_word[df_word["count"] >= 2]
df_word
```

```python
# 빈도표 만들기
df_word = (
    df_word.groupby("word", as_index=False)
    .agg(n=("word", "count"))
    .sort_values("n", ascending=False)
)
df_word
```

```python
top20 = df_word.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update(
    {
        "figure.figsize": [6.5, 6],
    }
)

plt.rc("font", family="NanumBarunGothic")

sns.barplot(data=top20, x="n", y="word", palette="hls")
```

```python
# 딕셔너리로 만들기
dic_word = df_word.set_index("word").to_dict()["n"]
dic_word
```

```python
from wordcloud import WordCloud

font = "D2Coding-Ver1.3.2-20180524-all.ttc"

wc = WordCloud(
    random_state=1, width=1000, height=1000, background_color="white", font_path=font
)

img_wordcloud = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(5, 5))
plt.axis("off")
plt.imshow(img_wordcloud)
```
